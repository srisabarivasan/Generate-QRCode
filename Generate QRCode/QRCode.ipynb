```python
import qrcode
```

```python
import image
```

```python
qr = qrcode.QRCode(version = 5, box_size = 5, border = 5)
```

```python
data = "https://www.linkedin.com/in/srisabarivasan"
```

```python
qr.add_data(data)
```

```python
qr.make(fit = True)
```

```python
img = qr.make_image(fill = "black", back_color= "white")
```

```python
img.save("QRCode.png")
```
